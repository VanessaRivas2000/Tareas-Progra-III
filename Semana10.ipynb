```python
# Pasar en variables los datos de entrenamiento y prueba
```

```python
datos_entrenamiento = metadata.splits ["train"].num_examples
datos_prueba = metadata.splits ["test"].num_examples
print("datos_entrenamiento", datos_entrenamiento, "datos_prueba", datos_prueba)
```

```python
# Funcion normalizadora de los pixeles de 0 - 255 que sean de 0 - 1
```

```python
def normalizacion(images, labels):
  images = tf.cast(images. tf.float32)
  images /= 255
  return images, labels
```

```python
# Llamamos la funcion normalizadora en ambas dataset
```

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28,1)),
    tf.keras.layers.Dense(64, activation = tf.nn.relu),
    tf.keras.layers.Dense(64, activation = tf.nn.relu),
    tf.keras.layers.Dense(10, activation = tf.nn.relu),

])
```

```python
modelo.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
# Entrenamiento por lotes
```

```python
batchize = 32
datos_entrenamiento = datos_entrenamiento.repeat().shuffle(datos_entrenamiento).batch(batchize)
datos_entrenamiento = datos_prueba.repeat().shuffle(datos_prueba).batch(batchize)
```
